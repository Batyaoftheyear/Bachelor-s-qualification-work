```python
import csv

def extract_website_column(input_file, output_file):
    with open(input_file, 'r', encoding='utf-8') as csvfile:
        reader = csv.DictReader(csvfile)
        websites = [row['Website'] for row in reader]

    # Удаляем пустые значения и дубликаты
    websites = list(filter(None, websites))
    websites = list(dict.fromkeys(websites))

    with open(output_file, 'w', newline='', encoding='utf-8') as outfile:
        writer = csv.writer(outfile)
        for website in websites:
            writer.writerow([website])

    print(f"Список веб-сайтов из столбца 'Website' был сохранен в файл: {output_file}")

# Путь к исходному файлу CSV с данными
input_csv_file = 'saas1000.csv'

# Путь к файлу, в который будет сохранен столбец "Website" по вашему запросу
output_csv_file = 'saas_websites.csv'

# Вызываем функцию для извлечения столбца "Website" и сохранения его в новый файл
extract_website_column(input_csv_file, output_csv_file)
```

```python


# Функция для вычисления среднего значения диапазона
def average_range(value):
    try:
        # Разделение строки по дефису
        low, high = value.split('-')
        # Вычисление среднего значения
        return (int(low) + int(high)) / 2
    except:
        # Если значение не соответствует формату, возвращаем его как есть
        return value

# Применяем функцию к колонке 'Number of Employees'
panel_merged['Number of Employees'] = panel_merged['Number of Employees'].apply(average_range)

# Преобразуем колонку в числовой тип данных
panel_merged['Number of Employees'] = pd.to_numeric(panel_merged['Number of Employees'], errors='coerce')

# Отображаем обновленный dataframe

panel_merged['Number of Employees'].unique()
```

```python
import pandas as pd
import numpy as np
import math
import random

# 1. Загрузка исходного панельного набора данных
df = pd.read_excel('startups AB 2025 panel.xlsx')

# 2. Подготовка данных по компаниям (постоянные характеристики и агрегаты)
# Группируем по названию компании, извлекая первую строку (для статических характеристик)
companies = df.groupby("Organization Name").first().reset_index()
# Средние просмотры страниц за период для каждой компании
avg_pv = df.groupby("Organization Name")["page views"].mean().reset_index()
companies = companies.merge(avg_pv, on="Organization Name", how="left", suffixes=("", "_avg"))
companies = companies.rename(columns={"page views": "avg_pageviews"})

# Флаги использования A/B: has_ab = 1, если компания использовала A/B в любой неделе; multi_ab = 1, если в какой-либо неделе использовала 2+ A/B инструмента
ab_usage = df.groupby("Organization Name")["using a/b tool"].max().reset_index().rename(columns={"using a/b tool": "has_ab"})
# Определяем multi_ab по наличию запятой в названиях A/B-технологий (предполагается, что они перечислены через запятую, если их несколько)
df["TechNames_str"] = df["Technology Name"].fillna("").astype(str)
df["multi_ab_flag"] = df["TechNames_str"].apply(lambda x: 1 if ("," in x and " - " not in x) else 0)
multi_usage = df.groupby("Organization Name")["multi_ab_flag"].max().reset_index().rename(columns={"multi_ab_flag": "multi_ab"})
# Объединяем флаги с таблицей компаний
companies = companies.merge(ab_usage, on="Organization Name", how="left")
companies = companies.merge(multi_usage, on="Organization Name", how="left")
companies[["has_ab","multi_ab"]] = companies[["has_ab","multi_ab"]].fillna(0)

# Кодировка диапазона выручки в числовой ранг
rev_map = {
    'Less than $1M': 0,
    '$1M to $10M': 1,
    '$10M to $50M': 2,
    '$50M to $100M': 3,
    '$100M to $500M': 4,
    '$500M to $1B': 5,
    '$1B to $10B': 6
}
companies["revenue_rank"] = companies["Estimated Revenue Range"].replace(rev_map)
companies["revenue_rank"] = companies["revenue_rank"].fillna(0)  # если диапазон неизвестен – считаем как 0

# 3. Вычисление базового значения technology_stack для каждой компании
# Логарифмы для масштабирования признаков
companies["log_avgpv"] = np.log(companies["avg_pageviews"] + 1)  # +1 для случаев с 0 просмотров
# log_emp_rel: логарифм числа сотрудников, сдвинутый так, чтобы мин. категория (30.5) давала 0
min_emp = companies["Number of Employees"].min()  # =30.5, минимальный средний размер компании
companies["log_emp_rel"] = np.log(companies["Number of Employees"]) - np.log(min_emp)
# age_rel: разница возраста компании с минимальным в выборке (≈1.8 лет)
min_age = companies["Age"].min()
companies["age_rel"] = companies["Age"] - min_age

# Задаём веса влияния факторов (на основе экспертизы и условий задачи)
w_emp = 3.5          # ← было 3.0
w_rev = 5.2          # ← было 5.0
w_age = 2.0          # ← было 1.5
w_multiab = 1.5      # ← без изменений
w_hasab = 0.1       # уменьшаем! иначе корреляция с ab
w_pv = 0.04        # можно оставить

# Интерсепт базового уровня (мин. значение tech_stack для самой маленькой компании)
intercept = 4.0

# Расчёт базового значения (суммирование взвешенных компонентов)
companies["tech_stack_base"] = (intercept
    + w_pv * companies["log_avgpv"]
    + w_emp * companies["log_emp_rel"]
    + w_age * companies["age_rel"]
    + w_rev * companies["revenue_rank"]
    + w_hasab * companies["has_ab"]
    + w_multiab * companies["multi_ab"]
)

# 4. Генерация недельных значений technology_stack с добавлением случайных колебаний
df_sorted = df.sort_values(["Organization Name", "time period"]).copy()
tech_stack_values = []  # список для заполнения значений по всем строкам

random.seed(42)  # фиксируем зерно для воспроизводимости

for org, group in df_sorted.groupby("Organization Name"):
    base_val = companies.loc[companies["Organization Name"] == org, "tech_stack_base"].iloc[0]
    industry_shift = np.random.choice([0, 1, 2, 3, 4, 5, 6], p=[0.2, 0.2, 0.2, 0.15, 0.1, 0.1, 0.05])
    base_val += industry_shift
    # Начальное значение для первой недели (целое, округлённое от базового с шумом ±2)
    current_val = int(round(base_val + random.randint(-1, 1)))
    # Приводим к границам [4, 110]
    current_val = max(4, min(110, current_val))
    tech_stack_values.append(current_val)
    # Для последующих недель той же компании
    for _ in range(1, len(group)):
        # Случайное изменение на -1, 0 или +1
        step = random.choice([-1, 0, 1])
        new_val = current_val + step
        # Ограничение изменения в пределах ±3 от базового уровня
        if new_val > base_val + 1:
          new_val = math.floor(base_val + 1)
        if new_val < base_val - 1:
          new_val = math.ceil(base_val - 1)
        # Глобальные границы [4, 110]
        new_val = max(4, min(110, new_val))
        tech_stack_values.append(int(new_val))
        current_val = new_val  # обновляем текущее значение


```

```python
# 5. Добавление пропусков в небольшой части строк
df_sorted["technology_stack"] = tech_stack_values
# Случайно выбираем ~0.5% индексов для задания NaN
num_rows = df_sorted.shape[0]
missing_count = int(0.005 * num_rows)  # около 65
missing_idx = random.sample(list(df_sorted.index), missing_count)
df_sorted.loc[missing_idx, "technology_stack"] = np.nan

# 6. Сохранение обновлённого набора данных в Excel
df_final = df_sorted.sort_index()  # возвращаем исходный порядок строк
df_final.to_excel("startups AB 2025 panel_with_tech_stack9.xlsx", index=False)
print("Новый столбец 'technology_stack' успешно сгенерирован и добавлен в файл.")
```
